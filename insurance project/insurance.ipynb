```python
##4.	Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns
```

```python
sns.pairplotot(x_vars='charges',y_vars=['age','bmi','children'],data=data1)

plt.show()
```

```python
##or
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
axes[0].scatter('charges','age',data=data1,color='red')
axes[0].set_title('age scatter plot')

axes[1].scatter('charges','bmi',data=data1,color='orange')
axes[1].set_title('bmi scatter plot')

axes[2].scatter('charges','children',data=data1,color='green')
axes[2].set_title('children scatter plot')

plt.show()
```

```python
categorical_cols=[data1['sex'],data1['smoker'],data1['region']]
```

```python
categorical_cols
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5))
for i,cols in enumerate(categorical_cols):
    sns.countplot(x=cols,data=data1,ax=axes[i])
    axes[i].set_title(f"the countplot for {cols}")
plt.show()
```
